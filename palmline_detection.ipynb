**Importing Libraries**

```
import cv2
from google.colab.patches import cv2_imshow
```

```
img = cv2.imread('palm.jpg')
```

**Original image**

```
cv2_imshow(img)
```

**Converting BGR to Gray**

```
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

Now we use **canny edge detector** to find the Palm Lines

```
edges = cv2.Canny(gray_img,40,55,apertureSize = 3)
cv2_imshow(edges)
```

Now we revert the colors so that recognized lines are to black

```
edges = cv2.bitwise_not(edges)
cv2_imshow(edges)
```

Blend the previous image, with the original image

```
cv2.imwrite('palmline.jpg',edges)
palmlines = cv2.imread('palmline.jpg')
```

```
img1 = cv2.addWeighted(palmlines, 0.3, img , 0.7 , 0)
cv2_imshow(img1)
```
